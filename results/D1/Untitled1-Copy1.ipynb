```python
trainer1 = Trainer(
    dataset='D1',
    hidden_dims=[30] * 8,
    act_fn=jax.nn.relu,
    study_name='30x8_relu',
    residual=False
)

trainer2 = Trainer(
    dataset='D1',
    hidden_dims=[30] * 4 + [24] * 4,
    act_fn=jax.nn.relu,
    study_name='30x4_24x4_relu',
    residual=False
)

trainer3 = Trainer(
    dataset='D1',
    hidden_dims=[30] * 4 + [18] * 4,
    act_fn=jax.nn.relu,
    study_name='30x4_18x4_relu',
    residual=False
)

trainer4 = Trainer(
    dataset='D1',
    hidden_dims=[30] * 4 + [12] * 4,
    act_fn=jax.nn.relu,
    study_name='30x4_12x4_relu',
    residual=False
)

trainer5 = Trainer(
    dataset='D1',
    hidden_dims=[30] * 4 + [10] * 4,
    act_fn=jax.nn.relu,
    study_name='30x4_10x4_relu',
    residual=False
)

Trainer.compare_graphs(
    trainers=[trainer1, trainer2, trainer3, trainer4, trainer5],
    title=r'30x4 + $d$x4 ReLU',
    legend=[r'$d=30$', r'$d=24$', r'$d=18$', r'$d=12$', r'$d=10$'],
    save_graph=True,
    filename='30x4_dx4_relu'
)
```
